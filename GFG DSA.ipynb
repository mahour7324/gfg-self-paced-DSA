### 2 find HCF 

```python
def pri(a,b):# using rucursion
    b = max(a,b)
    a = min(a,b)
    if b%a == 0:
        return a
    else :
        temp = b%a
        b = a
        a = temp
        return pri(a,b)
l = pri(12,18)
print(l)
    
```

### 3 to find palidrome (mirror image after swape)

```python
def rev(n):
    total = 0
    r = n%10
    while r>0:
        r =  n%10
        n = n//10
        total = total*10+r
    total = total//10
    return total
print(rev(123))
        

```

```python
def checkpalidrome():
    N = int(input("enter the no : "))
    a = True if N == rev(N) else False
    print(a)
checkpalidrome()
```

### 4 find fabonacci series

```python
def fabo(n):
    while n  >=0:
        if n==0 or n== 1:
            return n
        else :
            n =  fabo(n-1) + fabo(n-2)
            return n
print(fabo(11))
    
```

### 6 find foctorial of a given no

```python
def factorial(n): # recursive solution of 5!
    if n < 1:
        return 1
    else:
        return n*factorial(n-1)
l =  factorial(5)
print(l)
    
    
```

```python
def factorial2(n):  # simple iterative solution
    total = 1
    while n>=1:
        total = total*n
        n = n-1
    return total
print(factorial2(100))
```

### 6 find HCF

```python
# using recursion
def hcf(a,b):
    temp1 = a
    temp2 = b
    b = max(temp1, temp2)
    a = min(temp1, temp2)
    if b%a ==0: # -----------------
        return a
    else:
        a = b%a
        return hcf(a,b) #----------
print(hcf(15,12))
    
```

```python
# leet code medium level question rotate array
import numpy as np
array = list(range(1,8))
print("the original array : ", array)

n = 3
def ra(array, n):
    
    size = array.size
    newarray  = np.array[size]
    limit = size -n
    print("size of array is : ",size)
    i  = size-1 
    j = 0
    while i>=limit:
        element= array[i]
        newarray = newarray.push(element)
        i = i-1
    while j<limit:
        element = array[j]
        newarray = newarray.push(element)
    return  newarray
l = ra(array,n)
print(l)
    
    

```
