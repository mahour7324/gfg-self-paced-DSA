# Besic programs

```python
import random as r
def rand():
    j  = r.randrange(1,100)
    print(j )
    return j
```

#### 1 programs to add two no

```python
def addTwoNo(a,b):
    return a+b
print(addTwoNo(rand(),rand()))
```

#### 2 max of two numbers

```python
def maxOfTwo(a,b):
    return a if a>b else b
print(maxOfTwo(rand(),rand()))


```

```python
# or ------------------------------------------------------------
def maxOutOf(a,b,*c):
    return max(a,b,*c)
print("the max is : ",maxOutOf(rand(),rand(),rand(),rand()))
```

#### 3 program for factorial of a no

```python
# ---------------------iterative solution-------------------------
def factorial(n):
    i = 2
    total = 1
    while i<=n:
        total  = total*i
        i = i+1
    return total
print(factorial(rand()))
```

```python
# -----------------------recursive solution-----------------------
def factorail1(n):
    return 1 if n == 0 else n*factorail1(n-1)
print(factorail1(rand()))
```
